# Histogram

Histogram adalah representasi grafis distribusi intensitas warna dalam gambar. Teknik perbaikannya meliputi:

Histogram Equalization → Meningkatkan kontras gambar.


Histogram Stretching (Kontras Stretching) → Memperluas rentang intensitas untuk meningkatkan detail.

```python
#Mengimport Library
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Membaca gambar dalam grayscale
image = cv2.imread('Jeruk_Mandarin.jpg', cv2.IMREAD_GRAYSCALE)

# Menampilkan histogram
plt.figure()
plt.hist(image.ravel(), bins=256, range=[0,256], color='black')
plt.title('Histogram Asli')
plt.xlabel('Intensitas Pixel')
plt.ylabel('Frekuensi')
plt.show()
```

Penjelasan:

image.ravel() → Mengubah citra menjadi array 1D untuk histogram.

bins=256 → Jumlah level keabuan (0-255).

plt.hist() → Menampilkan histogram intensitas piksel.


# Histogram Equalization

**Histogram Equalization meratakan distribusi intensitas sehingga gambar memiliki kontras lebih baik.**

```python
from google.colab.patches import cv2_imshow  # Untuk menampilkan gambar di Colab
```

```python
# Membaca gambar dalam grayscale
image = cv2.imread('Jeruk_Mandarin.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Histogram Equalization
equalized = cv2.equalizeHist(image)
```

```python
# Menampilkan hasil
cv2_imshow(image)       # Gambar Asli
cv2_imshow(equalized)   # Gambar Setelah Equalization
```

```python
# Menampilkan histogram sebelum & sesudah
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.hist(image.ravel(), bins=256, range=[0,256], color='black')
plt.title('Histogram Asli')

plt.subplot(1,2,2)
plt.hist(equalized.ravel(), bins=256, range=[0,256], color='black')
plt.title('Histogram Setelah Equalization')

plt.show()
```

#  Histogram Stretching (Kontras Stretching)

```python
# Mencari nilai minimum dan maksimum piksel
min_pixel = np.min(image)
max_pixel = np.max(image)
```

```python
# Kontras Stretching Formula
stretched = ((image - min_pixel) * (255.0 / (max_pixel - min_pixel))).astype(np.uint8)
```

```python
# Menampilkan hasil
cv2_imshow(image)       # Gambar Asli
cv2_imshow(stretched)   # Gambar Setelah Stretching
```

```python
# Menampilkan histogram sebelum & sesudah
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.hist(image.ravel(), bins=256, range=[0,256], color='black')
plt.title('Histogram Asli')

plt.subplot(1,2,2)
plt.hist(stretched.ravel(), bins=256, range=[0,256], color='black')
plt.title('Histogram Setelah Stretching')

plt.show()
```
